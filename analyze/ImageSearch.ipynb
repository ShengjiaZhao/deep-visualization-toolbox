```python
import matplotlib
matplotlib.use('Agg')
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from scipy import misc
```

```python
activation = np.load('/home/ubuntu/sdf/activations/fc6.npy')
num_image = activation.shape[0]
```

```python
# Read in path for all the images
image_root = '/home/ubuntu/sdf/images/'
image_list_file = open('/home/ubuntu/sdf/activations/input.txt')
image_list = []
while True:
    path = image_list_file.readline()
    if not path:
        break
    image_list.append(path.strip())
```

```python
import random
query = random.randint(0, num_image-1)
diff_matrix = activation - np.tile(activation[query, :], (num_image, 1))
result = np.argsort(np.sum(abs(diff_matrix), 1))
```

```python
for i in range(0, 10):
    plt.imshow(misc.imread(image_root + image_list[result[i]]))
    plt.show()
```
